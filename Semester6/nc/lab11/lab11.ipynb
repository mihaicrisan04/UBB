```python
import numpy as np
import sys


def integrate(fn, a, b, steps=5, debug=False, exact=None):
    table = np.zeros((steps, steps), dtype=np.float64)
    pow_4 = 4 ** np.arange(steps, dtype=np.float64) - 1
    n_evals = 2
    # trapezoidal rule
    h = (b - a)
    table[0, 0] = h * (fn(a) + fn(b)) / 2

    for j in range(1, steps):
        h /= 2
        # extended trapezoidal rule
        table[j, 0] = table[j - 1, 0] / 2
        table[j, 0] += h * sum(
            fn(a + i * h) for i in range(1, 2 ** j + 1, 2)
        )
        n_evals += 2 ** (j - 1)
        # richardson extrapolation
        for k in range(1, j + 1):
            table[j, k] = table[j, k - 1] + \
                (table[j, k - 1] - table[j - 1, k - 1]) / pow_4[k]
    # debug
    if debug:
        print()
        print(table, file=sys.stderr)
        if exact is not None:
            errors = [
                '%.2e' % i
                for i in np.abs(table.diagonal() - exact)
            ]
            print('abs. error:', errors, file=sys.stderr)
    return table[-1, -1], n_evals
```

```python
exact = np.pi * np.sqrt(3) / 9
f = lambda x: 1 / (2 + np.sin(x))

for steps in range(2, 15):
    approx, n_evals = integrate(f, 0, np.pi / 2, steps=steps, debug=True, exact=exact)
    if abs(approx - exact) < 0.5e-5:
        break

print(f"steps = {steps}")
print(f"approximation = {approx:.7f}")
print(f"exact (pi*sqrt(3)/9) = {exact:.7f}")
print(f"error = {abs(approx - exact):.3e}")
print(f"function evaluations = {n_evals}")
```
