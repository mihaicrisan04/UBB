## Problem 2: Using  U(0, 1)  to Simulate Distributions

We need to simulate four distributions:
1. **Bernoulli Distribution** *Bern(p)*
2. 	**Binomial Distribution**  *Bino(n, p)*
3. 	**Geometric Distribution** *Geo(p)*
4.	**Pascal (Negative Binomial) Distribution** *NB(n, p)*

```python
import numpy as np
import matplotlib.pyplot as plt

# Bernoulli Distribution
def simulate_bernoulli(p, size=1000):
    return np.random.choice([0, 1], size=size, p=[1-p, p])

# Binomial Distribution
def simulate_binomial(n, p, size=1000):
    return np.random.binomial(n, p, size=size)

# Geometric Distribution
def simulate_geometric(p, size=1000):
    return np.random.geometric(p, size=size) - 1  # Adjust to match the problem's definition (k failures)

# Pascal (Negative Binomial) Distribution
def simulate_pascal(n, p, size=1000):
    return np.random.negative_binomial(n, 1-p, size=size)

# Simulate and visualize
size = 1000
p = 0.5
n = 10

# Simulate distributions
bernoulli_data = simulate_bernoulli(p, size)
binomial_data = simulate_binomial(n, p, size)
geometric_data = simulate_geometric(p, size)
pascal_data = simulate_pascal(n, p, size)

# Plot the results
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].hist(bernoulli_data, bins=2, density=True, color='blue', alpha=0.7)
axs[0, 0].set_title('Bernoulli Distribution')
axs[0, 1].hist(binomial_data, bins=n+1, density=True, color='green', alpha=0.7)
axs[0, 1].set_title('Binomial Distribution')
axs[1, 0].hist(geometric_data, bins=20, density=True, color='purple', alpha=0.7)
axs[1, 0].set_title('Geometric Distribution')
axs[1, 1].hist(pascal_data, bins=20, density=True, color='orange', alpha=0.7)
axs[1, 1].set_title('Pascal (Negative Binomial) Distribution')

plt.tight_layout()
plt.show()
```

The histograms illustrate the distribution of simulated values.<br> For example, a Bernoulli distribution results in two bars (0 and 1),<br> while a binomial distribution has bars for  *0, 1, ..., n*.
