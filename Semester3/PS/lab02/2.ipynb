```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters
n = 3  # Number of trials
p = 0.5  # Probability of success
```

```python
# a) Probability Distribution Function (PDF)
k_values = np.arange(0, n+1)
pdf_values = stats.binom.pmf(k_values, n, p)

# Display PDF
print("PDF values:", pdf_values)
```

```python
# b) Cumulative Distribution Function (CDF)
cdf_values = stats.binom.cdf(k_values, n, p)

# Display CDF
print("CDF values:", cdf_values)
```

```python
# c) Specific probabilities
P_X_0 = stats.binom.pmf(0, n, p)
P_X_not_1 = 1 - stats.binom.pmf(1, n, p)

print("P(X = 0):", P_X_0)
print("P(X ≠ 1):", P_X_not_1)
```

```python
# d) Cumulative probabilities
P_X_leq_2 = stats.binom.cdf(2, n, p)
P_X_less_2 = stats.binom.cdf(1, n, p)  # P(X < 2) = P(X <= 1)

print("P(X ≤ 2):", P_X_leq_2)
print("P(X < 2):", P_X_less_2)
```

```python
# e) Complementary probabilities
P_X_geq_1 = 1 - stats.binom.cdf(0, n, p)  # P(X ≥ 1) = 1 - P(X < 1)
P_X_greater_1 = 1 - stats.binom.cdf(1, n, p)  # P(X > 1) = 1 - P(X ≤ 1)

print("P(X ≥ 1):", P_X_geq_1)
print("P(X > 1):", P_X_greater_1)
```

```python
# f) Simulate 3 coin tosses
np.random.seed(42)  # For reproducibility
simulated_tosses = np.random.binomial(n, p, size=10)  # Simulate 10 trials
print("Simulated coin toss results:", simulated_tosses)
```

```python
# Plot PDF and CDF
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.bar(k_values, pdf_values, color='blue', alpha=0.7)
plt.title("Probability Distribution Function (PDF)")
plt.xlabel("Number of Heads (X)")
plt.ylabel("Probability")

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(k_values, cdf_values, where='post', color='orange', alpha=0.7)
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Number of Heads (X)")
plt.ylabel("Cumulative Probability")

plt.tight_layout()
plt.show()
```
